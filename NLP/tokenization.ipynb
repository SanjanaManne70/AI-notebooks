```python
#customizing tokenization
from spacy.symbols import ORTH

nlp = spacy.blank("en")
doc = nlp("gimme double cheese extra large healthy pizza")
tokens = [token.text for token in doc]
tokens
```

```python

nlp.tokenizer.add_special_case("gimme", [
    {ORTH: "give"}, #but we can't modify the word to "give" like that if tried:error
    {ORTH: "me"},
])
doc = nlp("gimme double cheese extra large healthy pizza")
tokens = [token.text for token in doc]
tokens
```

```python
nlp.tokenizer.add_special_case("gimme", [
    {ORTH: "gim"}, #but we can't modify the word to "give" like that if tried:error
    {ORTH: "me"},
])
doc = nlp("gimme double cheese extra large healthy pizza")
tokens = [token.text for token in doc]
tokens
```

```python
#Using PIPELINE in nlp
#download trained pipeline using
#python -m spacy download en_core_web_sm
nlp= spacy.load("en_core_web_sm") #en:english, sm: small
#the pipeline contains in-built features
```

```python
nlp.pipeline
```

```python
nlp.pipe_names
```

```python
doc= nlp("Tony gave two $ to peter.")
for token in doc:
    print(token, "|", token.pos_, "|", token.lemma_)
    #pos:parts of speech, lemma: base form
    #for blanl spacy pipeline it wll be blank
```

```python
#note:few local languages like hindi does not have pipelines in spacy

doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents: #ent=entity
    print(ent.text,"|", ent.label_, "|", spacy.explain(ent.label_))
```

```python
nlp= spacy.blank("en")
doc= nlp("Tesla Inc is going to acquire twitter for $45 billion")
for token in doc:
    print(token,"|", token.pos_,"|", token.lemma_)
```

```python
#ADDING A COMPONENT TO A BLANK PIPELINE
#from english trained pipeline add NER component
source_nlp= spacy.load("en_core_web_sm")
nlp= spacy.blank("en")
nlp.add_pipe("ner", source= source_nlp)
nlp.pipe_names
```
