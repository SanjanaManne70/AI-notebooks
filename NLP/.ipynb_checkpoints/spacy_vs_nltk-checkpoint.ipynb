```python
#pip install nltk
# pip install spacy
#python -m spacy download en ..en:english
import spacy
```

```python
#spacy is OBJECT -ORIENTED
nlp= spacy.load("en_core_web_sm")
doc= nlp("Mr.Sanju love naruto.My brother loves telugu movies")
for sentence in doc.sents:
    print(sentence)
```

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```

```python
from nltk.tokenize import sent_tokenize

text = "Mr.Sanju love naruto. My brother loves telugu movies"
print(sent_tokenize(text))
```

```python
from nltk.tokenize import word_tokenize

word_tokenize("Mr.Sanju love naruto. My brother loves telugu movies")
```
