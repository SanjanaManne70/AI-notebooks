```python
#Stemming:Cuts off word endings (prefixes/suffixes) using rules
#running → run
#studies → studi (not a valid word)
#Cons: Can produce non-meaningful words.
#...........
#Lemmatizing: Uses vocabulary + grammar rules to
#find the proper base form (lemma).
#studies → study
#.........
#Spacy supports only lemmatization
#NLTK supports both stemming and lemmatizing
import spacy
import nltk
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
words = ["eating", "eats", "eat", "ate", "adjustable", "rafting", "ability", "meeting"]

for word in words:
    print(word, "|", stemmer.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

words = ["eating", "eats", "eat", "ate", "adjustable", "rafting", "ability", "meeting"]

for word in words:
    # Default POS is noun, so verbs like "ate" need pos="v"
    print(word, "|", lemmatizer.lemmatize(word, pos="v"))
#PorterStemmer: Rule-based, fast, sometimes messy (abil).
#WordNetLemmatizer: Dictionary-based, 
```

```python
#Lemmatization in spacy!!
nlp = spacy.load("en_core_web_sm")
doc= nlp("eating eats eat ate adjustable rafting ability meeting better")
for token in doc:
    print(token, "|" ,token.lemma_)
```

```python
#CUSTOMIZING SPACY LEMMATIZER: since  default lemmatizer doesn’t know slang like bro or brah.
#note: NLTK- WordNetLemmatizer-not easily customizable
#in spacy_pipeline: attributer_ruler is a pipeline component that lets u add custom attributes(Like "lemma", "pos", etc)
ar = nlp.get_pipe('attribute_ruler')

ar.add([[{"TEXT":"Bro"}],[{"TEXT":"Brah"}]],{"LEMMA":"Brother"})

doc = nlp("Bro, you wanna go? Brah, don't say no! I am exhausted")
for token in doc:
    print(token.text, "|", token.lemma_)
```
