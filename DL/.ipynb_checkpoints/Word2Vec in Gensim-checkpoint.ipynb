```python
import pandas as pd 
import gensim
```

```python
#the dataset we're using is a subset of amazon reviews from the 
#cell phones and accessories category
df= pd.read_json("Cell_Phones_and_Accessories_5.json", lines=True)
df
```

```python
df.shape
```

```python
#Data preprocessing
review_text= df.reviewText.apply(gensim.utils.simple_preprocess)
review_text
#tokenization is done and removed words like a, an, I, etc.
```

```python
#loc[] is a index label-based indexer in pandas
#used to select rows and/or columuns by their labels not position index
#while iloc[]  is integer position based 
review_text.loc[0]
```

```python
df.reviewText.loc[0]
```

```python
#Train the model for reviews. 
#Use a window of size 10 -10 words before the present word and 10words ahead.
#A sentence with at least 2 words should only be considered- min_count
#Workers define how many CPU threads to be used.


#initialize the model 
model= gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers= 4,
)
```

```python
#Build Vocabulary
model.build_vocab(review_text, progress_per=1000)
```

```python
#train model
model.train(review_text, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
#Save the model so tat it can be reused in other applications
model.save("./word2vec-amazon-cell-accessories-reviews-short.model")
```

```python
#Finding similar words and similarity between words
model.wv.most_similar("bad")
```

```python
#we can pint cosine similarity score between 2  given words also
model.wv.similarity(w1="cheap",w2="inexpensive")
```

```python
model.wv.similarity(w1="great",w2="chips")
```
