```python
#using hot one encoder
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
le=LabelEncoder()
```

```python
dfle['town '] = le.fit_transform(dfle['town '])
```

```python
X = dfle[['town ', 'area']].values  # 2D array with two columns
Y = dfle['price'].values
```

```python
ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(drop='first'), [0])  # drop='first' avoids dummy variable trap
    ],
    remainder='passthrough'  # keep 'area' as is
)
```

```python
X = ct.fit_transform(X)
```

```python
# Fit linear regression model
model = LinearRegression()
model.fit(X, Y)
```

```python
prediction = model.predict([[0, 0, 3000]])
print(prediction)
```
