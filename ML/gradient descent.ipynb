```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
def gradient_descent(x,y):
    w_curr= b_curr=0
    iterations=100
    n= len(x)
    
    learning_rate= 0.01
    cost_list = []  # store cost per iteration
    w_list = []  
    for i in range(iterations):
        y_predicted = w_curr*x+ b_curr
        cost= (1/n)* sum([val**2 for val in (y-y_predicted)])
        wd= -(2/n)* sum(x*(y-y_predicted))
        bd= -(2/n)*sum(y-y_predicted)
        w_curr= w_curr-learning_rate*wd
        b_curr= b_curr-learning_rate*bd
        print("w {} , b {} ,cost {} itertion {}".format(w_curr,b_curr,cost,i))   

    return w_curr, b_curr, cost_list, w_list
    
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
w_final, b_final, cost_list, w_list = gradient_descent(x,y)
```

```python
w_final, b_final, cost_list, w_list = gradient_descent(x,y)

# Plot Regression Line
#plt.figure(figsize=(8,5))
plt.scatter(x,y, color='red', label='Actual points')
y_predicted = w_final*x + b_final
plt.plot(x, y_predicted, color='blue', label='Regression line')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression using Gradient Descent")
plt.legend()
plt.show()
```
