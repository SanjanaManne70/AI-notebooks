```python
import pandas as pd
import matplotlib.pyplot as plt# or  from matplotliv import pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv("insurance.csv")
df
```

```python
plt.scatter(df['age'],df['bought_insurance'],marker='+',color='orange')
```

```python
#df.shape  gives the number of rows and columns in the csv
df.shape
```

```python
from sklearn.model_selection import train_test_split
#to split the dataset into training and testing sets..
X = df[['age']]               # Features (input)
y = df['bought_insurance']  
```

```python
#hey if u click on "SHIFT+ TAB" u get help dialgog for arguments related stuf! 

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model= LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test) # first value is prob of class 0 and nest is prob for 1
```

```python
model.predict([[93]])
```

```python
#if we have less score , then use CONFUSION MATRIX to understand and improve your model 
y_predicted= model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm= confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
