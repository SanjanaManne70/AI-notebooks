```python
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits
```

```python
digits= load_digits()
```

```python
dir(digits)
# dir(object_name) returns the LIST OF ALL ATTRIBUTES AND METHODS OF AN OBJECT  
```

```python
digits.data[0]
```

```python
plt.gray()
plt.matshow(digits.images[0])#first image in digits datset displayed as a matrix(2d array)
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
 from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(digits['data'],digits.target,test_size=0.2)
```

```python
#len(X_train)
X_train
```

```python
from sklearn.linear_model import LogisticRegression

```

```python
model= LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
model.predict([digits.data[67]])
```

```python
model.predict(digits.data[2:5])
```

```python
y_predicted= model.predict(X_test)
from sklearn.metrics import confusion_matrix
#confusion matrix is a visualization of how well ur model is doing
cm= confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')

#28 times truth was 0 and predicted also matched ....
but at truth 8 '5' times model predicted it wo=rongly as 1....
meaning in black area if u see number other than 0 there model is not working right
```
