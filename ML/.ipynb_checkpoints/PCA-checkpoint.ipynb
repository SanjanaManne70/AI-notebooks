```python
dataset.target[9] #so datasset.data is feature and dataset.
#target is the class that we obtain after classification
```

```python
#craeting dataframe
df= pd.DataFrame(dataset.data,columns=dataset.feature_names)
df.head()
```

```python
df.describe()
#min is 0
#max is 16
```

```python
x= df
y= dataset.target
```

```python
y
```

```python
#before training the model ,scaling karo
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_scaled=scaler.fit_transform(x)
X_scaled #we can use minmaxscaler also
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y, test_size=0.2,random_state= 30)
```

```python
#ready to train the model yo!
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
#do PCA now...
from sklearn.decomposition import PCA
#pca= PCA(n_components=2) or
pca= PCA(0.95) #meaning retain/capture  95% useful info and features uk

x_pca = pca.fit_transform(x)# new "principle component"
x_pca.shape

```

```python
pca.explained_variance_ratio_ # see in first column it only captured 14 % useful info
```

```python
pca.n_components_ # see no.of columns reduced from 64 to 29
```

```python
#so after PCA u can use the new Dtaframe to train your model
X_train_pca, X_test_pca, y_train,y_test=train_test_split(x_pca,y,test_size=0.2,random_state=30)
```

```python
model= LogisticRegression()
model.fit(X_train_pca,y_train)
model.score(X_test_pca,y_test)
```

```python
model= LogisticRegression(max_iter= 1000)

model.fit(X_train_pca,y_train)
model.score(X_test_pca,y_test)
#see 
```

```python
pca= PCA(n_components= 2)#it will identify the two most important features
x_pca= pca.fit_transform(x)
x_pca.shape
#computation is fast but accuracy reduces in PCA
```

```python
x_pca
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train,y_test=train_test_split(x_pca,y,test_size=0.2,random_state=30)
model= LogisticRegression(max_iter= 1000)

model.fit(X_train_pca,y_train)
model.score(X_test_pca,y_test)
```
