```python
import pandas as pd
from sklearn.datasets import load_digits
digits= load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])#its images not image note!!
```

```python
df =pd.DataFrame(digits.data)
df.head()
```

```python
digits.data[:5]
```

```python
digits.target
```

```python
df['target']= digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split #(x,y,size)
X_train,X_test,y_train,y_test= train_test_split(df.drop(['target'],axis='columns'),digits.target,test_size=0.2)
```

```python
len(X_test)
```

```python
#here we use multiple algo to predict the outcome(decision trees) so use ensemble
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier(n_estimators=120)
# tuning the score by changing no of decision trees used for predicting
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_predicted= model.predict(X_test)
# truth on y-axis, and predicted on x_axis
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,y_predicted)
cm
```

```python
#for more visualization 
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
